```python
from nsepy import get_history
from datetime import date
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import cufflinks as cf
import seaborn as sns
%matplotlib inline

meag_cap  = pd.read_csv("mega_cap.csv")
Tickers= meag_cap['Symbol'].tolist()
print(len(Tickers))
```

# 7. crating two file one "stock.ns" for yahoo and just "stock" for nse 

```python
NS = []
BO = []
NSE = []
for sym in Tickers:
    x = sym.split('.')
    if x[1] == 'NS':
        #print (sym)
        NS.append(sym)
        NSE.append(x[0])
    if x[1] == 'BO':
        BO.append(sym)
    else:
        pass
#print(NS)
print("NS stocks :" ,len(NS))
#print(BO)
print("BO stocks :" ,len(BO))
total = len(NS) + len(BO)
print("Total stocks :" , total)
```

```python
NS = pd.DataFrame(data=NS)
NS.columns =['Symbol']
NS.to_csv("meg_cap_NS.csv", sep=',', index = False)
NSE = pd.DataFrame(data=NSE)
NSE.columns =['Symbol']
NSE.to_csv("meg_cap_NSE.csv", sep=',', index = False)
```
