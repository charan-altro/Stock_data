# Read the data as time range

## Method 1

```python
import pandas as pd
import numpy as np
stock = pd.read_csv("All_stocks_yfinance/TCS.NS.csv" , index_col = "Date" , squeeze = True)
Privous_Year = stock.loc['2020-01-01':'2020-12-31']
Privous_Year
```

## Method 2

```python
from datetime import datetime
# import datetime
# datetime.datetime.strptime
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d')
df = pd.read_csv("All_stocks_yfinance/TCS.NS.csv", parse_dates=['Date'], date_parser=dateparse)
Privous_Year = df[(df['Date']>pd.Timestamp(2020,1,1)) & (df['Date']<pd.Timestamp(2020,12,31))]
Privous_Year
```

## Method 3

```python
from datetime import date
import pandas as pd
from datetime import datetime
# import datetime
# datetime.datetime.strptime
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d')
df = pd.read_csv("All_stocks_yfinance/TCS.NS.csv", parse_dates=['Date'], date_parser=dateparse)
value_to_check = pd.Timestamp(date.today().year - 2, 12, 31)  
filter_mask = df['Date'] > value_to_check 
filtered_df = df[filter_mask]
filtered_df
```
